## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Create the bins in which Data will be held
# Bins are 0, 59.9, 69.9, 79.9, 89.9, 100.   
bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 100]

# Create the names for the five bins
age_categories = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"] 

unique_df = purchase_data.drop_duplicates(subset=["SN"])
unique_df["Age Ranges"] = pd.cut(unique_df["Age"], bins, labels=age_categories, include_lowest=False)
unique_df.head()
```

```python
age_group_df = unique_df.groupby("Age Ranges")
total_count_by_age = age_group_df["SN"].count()
percent_by_age = (age_group_df["SN"].count()/total_players)*100
age_demo_df = pd.DataFrame({"Age Ranges": age_categories, "Total Count": total_count_by_age, "Percentage of Players": percent_by_age})
age_demo_df["Percentage of Players"] = age_demo_df["Percentage of Players"].map("{:.2f}%".format)
age_demo_df.set_index("Age Ranges", inplace=True)
age_demo_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python

purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"], bins, labels=age_categories, include_lowest=False)
purchase_data.head()
count_df = purchase_data.groupby("Age Ranges")["SN"].count()
count_df
avg_purch_price_df = purchase_data.groupby("Age Ranges")["Price"].mean()
avg_purch_price_df
total_purch_df = purchase_data.groupby("Age Ranges")["Price"].sum()
total_purch_df
purch_analysis_by_age = pd.merge(count_df, avg_purch_price_df, on="Age Ranges")


purch_analysis_by_age = pd.merge(purch_analysis_by_age, total_purch_df, on="Age Ranges")
purch_analysis_by_age.rename(columns={"SN":"Purchase Count", "Price_x":"Average Purchase Price", "Price_y": "Total Purchase Value"}, inplace=True)
purch_analysis_by_age["Avg Total Purchase per Person"] = purch_analysis_by_age["Total Purchase Value"]/age_demo_df["Total Count"]
purch_analysis_by_age["Average Purchase Price"] = purch_analysis_by_age["Average Purchase Price"].map("${:.2f}".format)
purch_analysis_by_age["Total Purchase Value"] = purch_analysis_by_age["Total Purchase Value"].map("${:.2f}".format)
purch_analysis_by_age["Avg Total Purchase per Person"] = purch_analysis_by_age["Avg Total Purchase per Person"].map("${:.2f}".format)
purch_analysis_by_age
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
num_purch = purchase_data.groupby("SN")["Purchase ID"].count()
total_purch = purchase_data.groupby("SN")["Price"].sum()

top_spenders = pd.merge(num_purch, total_purch, on="SN")
top_spenders.rename(columns={"Purchase ID": "Purchase Count", "Price": "Total Purchase Value"}, inplace=True)
top_spenders["Average Purchase Price"] = top_spenders["Total Purchase Value"]/top_spenders["Purchase Count"]
top_spenders["Average Purchase Price"] = top_spenders["Average Purchase Price"].map("${:.2f}".format)
top_spenders["Total Purchase Value"] = top_spenders["Total Purchase Value"].map("${:.2f}".format)

cols = ["Purchase Count", "Average Purchase Price", "Total Purchase Value"]
top_spenders = top_spenders[cols]
top_spenders = top_spenders.sort_values("Total Purchase Value", ascending=False)
top_spenders
```
